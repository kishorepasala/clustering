# K-MEAN 

```python
crime1 = pd.read_csv('crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_func(crime1.iloc[:,1:])
```

```python
df_norm
```

```python
# Elbow curv
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, n_init = 'auto')
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model = KMeans(n_clusters = 4)
model.fit(df_norm)
model.labels_
```

```python
x = pd.Series(model.labels_)
crime1['Clust'] = x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

# DBSCAN

```python
crime3 = pd.read_csv('crime_data.csv')
```

```python
crime3.info()
```

```python
df = crime3.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler = StandardScaler().fit(df.values)
x = stscaler.transform(df.values)
```

```python
x
```

```python
dbscan = DBSCAN(eps = 1,min_samples = 5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns = ['cluster'])
```

```python
cl
```

```python
pd.concat([crime3,cl],axis=1)
```

```python
# Adding clusters to dataset
crime['clusters'] = dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize = (10, 7))  
plt.scatter(crime['clusters'],crime['UrbanPop'], c = dbscan.labels_) 
```
